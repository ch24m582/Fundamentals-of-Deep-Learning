```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,3,6,8,9,10])
y = np.array([5,7.3,15.2,19.5,20.5,23])
```

```python
def batch_grad_descent(x,y,iterations,learning_rate):
    m =len(y)
    w = np.array([0.0,0.0], dtype = float)
    losses = []
    for i in range(iterations):
        y_pred = w[0] + w[1]*x
        err = y_pred -y
        loss = (1/2*m)*np.sum(err**2)
        losses.append(loss)
        grad0 = (1/m)*np.sum(err)
        grad1 = (1/m)*np.sum(err*x)
        w[0] -= learning_rate*grad0
        w[1] -= learning_rate*grad1
    return w,losses
```

```python
w , losses = batch_grad_descent(x,y,1000,0.04)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data points')

# Add the regression line
y_pred = w[0] + w[1]*x
plt.plot(x, y_pred, color='red', label='Regression Line')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Data with Regression Line: y vs x')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(losses, color='red', label='Loss over iterations')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss Function during Gradient Descent')
plt.grid(True)
plt.legend()
plt.show()
```
